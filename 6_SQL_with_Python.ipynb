Dies ist ein Notebook, mit dem wir SQL-Statements aus Python heraus ausführen wollen.

```python
#Welche Module benötigen wir, um mit der Datenbank zu sprechen?
import mysql.connector                #Treiber
import pandas as pd                   #Pandas halt... ;-)
from sqlalchemy import create_engine  #Verbindung zur Datenbank
```

```python
#Definieren des Connection-Strings
str_db_typ = 'mysql+mysqlconnector://' #Treiberinformationen
str_db_user = 'root'                   #Benutzer, der auf die Datenbank zugreift
str_db_password = ''                   #Passwort des Benutzers, der auf die Datenbank zugreift. 
str_db_adr = '@localhost'              #Adresse der Datenbank im Netzwerk
str_db_schema = '/northwind'           #Der Name der Datenbank, mit der wir standardmäßig arbeiten wollen.
```

```python
#%whos
```

```python
#Kombinieren eines String-Objektes zu einem 'Connection-String'
str_connection = str_db_typ + str_db_user + str_db_password + str_db_adr + str_db_schema
```

```python
str_connection
```

```python
#Verbindung zur Datenbank aufbauen
engine = create_engine(str_connection)
```

```python
# Erster einfacher Test, ob die Verbindung mit der Datenbank möglich war:
str_sql = 'SHOW TABLES'
pd.read_sql(str_sql, engine)
```

```python
### Alle Abfragen, die ein Ergebnis zurück liefern, können jetzt mit den nachfolgenden beiden Zeilen ausgeführt werden.
str_sql = 'HIER SQL STATEMENT EINTRAGEN'
pd.read_sql(str_sql, engine)
```

```python
#Zeige mir alles von...
str_sql = 'SELECT * FROM Products'
pd.read_sql(str_sql, engine)
```

```python
#Gezieltes Abfragen von einzelnen Spalten
str_sql = 'SELECT ContactName, City FROM Customers'
pd.read_sql(str_sql, engine)
```

```python
#Abfragen ohne und mit eindeutigen Werten (DISTINCT)
str_sql = 'SELECT Country FROM Customers'
pd.read_sql(str_sql, engine)
```

```python
str_sql = 'SELECT DISTINCT Country FROM Customers'
pd.read_sql(str_sql, engine)
```

```python
str_sql = 'SELECT COUNT(DISTINCT Country) FROM Customers'
pd.read_sql(str_sql, engine)
```

```python
# Abfragen von Daten, die bestimmte Bedingungen erfüllen
str_sql = "SELECT * FROM Customers WHERE Country='Mexico'"
pd.read_sql(str_sql, engine)
```

```python
#Abfragen von Daten, die mehreren Bedingungen genügen
str_sql = "SELECT * FROM Customers WHERE Country='Germany' AND City='Berlin'"
pd.read_sql(str_sql, engine)
```

```python
str_sql = "SELECT * FROM Customers WHERE (Country='Germany' OR Country='Spain') AND City='Madrid'"
pd.read_sql(str_sql, engine)
```

```python
#Daten direkt in der Abfrage sortieren lassen
str_sql = "SELECT * FROM Customers ORDER BY Country DESC"
pd.read_sql(str_sql, engine)
```

```python
#Wie finde ich fehlende Werte?
str_sql = "SELECT ContactName, Address FROM Customers WHERE Region IS NULL"
pd.read_sql(str_sql, engine)
```

```python
#Wie bekomme ich die ersten x Einträge aus einer Tabelle?
str_sql = "SELECT * FROM Customers ORDER BY Country DESC LIMIT 3 "
pd.read_sql(str_sql, engine)
```

```python
### Was tun, wenn ich nicht ganz genau weiß, wonach ich suche?
str_sql = "SELECT * FROM Customers WHERE CompanyName LIKE '%ant'" 
pd.read_sql(str_sql, engine)
```

```python
#Was ist denn, wenn ich mehrere Kriterien auf einmal in einer Spalte überprüfen möchte?
str_sql = "SELECT * FROM Customers WHERE Country IN ('Germany', 'France')"
pd.read_sql(str_sql, engine)
```

```python
### Wie kann man denn Intervalle abfragen?
str_sql = "SELECT * FROM Products WHERE UnitPrice BETWEEN 10 AND 20"
pd.read_sql(str_sql, engine)
```

```python
### Könnrn wir Daten auch direkt gruppiert abfragen?
str_sql = "SELECT COUNT(ProductID) AS Anzahl, AVG(UnitPrice) AS Mittelwert, ProductName FROM Products GROUP BY SupplierID"
pd.read_sql(str_sql, engine)
```

```python
### Könnrn wir Daten auch direkt gruppiert abfragen?
str_sql = "SELECT COUNT(CustomerID), Country FROM Customers GROUP BY Country"
pd.read_sql(str_sql, engine)
```

```python
## Wie kann ich denn Daten filtern, die ich schon aggregiert habe? --> Having
str_sql = "SELECT COUNT(CustomerID), Country FROM Customers \
GROUP BY Country HAVING COUNT(CustomerID) > 5 ORDER BY COUNT(CustomerID) DESC"
pd.read_sql(str_sql, engine)
```

```python
### Gibt es denn auch soetwas wie if-then-else?
str_sql = '''SHOW TABLES'''
pd.read_sql(str_sql, engine)
```

```python
### Gibt es denn auch soetwas wie if-then-else?
str_sql = '''SELECT OrderID, Quantity,
CASE 
WHEN Quantity > 30 THEN 'The quantity is greater than 30'
WHEN Quantity = 30 THEN 'The quantity is 30'
ELSE 'The quantity is under 30'
END AS QuantityText
FROM `Order Details`'''
pd.read_sql(str_sql, engine)
```

### Wie verknüpfen wir Tabellen bei der Abfrage?

```python
#LEFT JOIN
str_sql = '''SELECT A.CompanyName, A.PostalCode, B.OrderID
FROM Customers A
LEFT JOIN Orders B ON A.CustomerID = B.CustomerID
ORDER BY A.CompanyName'''
pd.read_sql(str_sql, engine)
```

```python
### INNER JOIN
str_sql = '''SELECT Orders.OrderID, Customers.CompanyName
FROM Orders
INNER JOIN Customers ON Orders.CustomerID = Customers.CustomerID'''
pd.read_sql(str_sql, engine)
```

```python
str_sql = '''SELECT * FROM Orders LIMIT 1'''
pd.read_sql(str_sql, engine)
```

### Wie kann ich denn Einträge wieder aus einer Tabelle entfernen?

```python
str_sql = "DELETE FROM CUSTOMERS WHERE CompanyName = 'Cardinal'"
engine.execute(str_sql)
```

### Wie kann ich denn Werte "überschreiben"?

### Wie können wir denn neue Objekte anlegen?
