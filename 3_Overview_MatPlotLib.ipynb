```python
# Styles are;

import math
x = np.arange(10)
n = len(plt.style.available)
num_rows = math.ceil(n/4)

fig = plt.figure(figsize=(15, 15))

for i, s in enumerate(plt.style.available):
    with plt.style.context(s):
        ax = fig.add_subplot(num_rows, 4, i+1)
        for i in range(1, 4):
            ax.plot(x, i * x**2, label='Group %d' % i)
            ax.set_xlabel(s, color='black')
            ax.legend(loc='best')
    
fig.tight_layout()
plt.show()
```

## Barplot

```python
divisions = ["Div-A", "Div-B", "Div-C", "Div-D", "Div-E"]
verkaefe = np.array([70, 80, 73, 65, 68])

plt.bar(divisions, verkaefe, color='green')
plt.title("BarPlot Example")
plt.xlabel("Division")
plt.ylabel("Sale")
plt.show()
```

```python
# Horizontal Barplot

divisions = ["Div-A", "Div-B", "Div-C", "Div-D", "Div-E"]
verkaefe = [70, 80, 73, 65, 68]
variance = [5,8,7,6,4]

plt.barh(divisions, verkaefe, xerr=variance, color='green')
plt.title("Horizontal BarPlot")
plt.ylabel("Division")
plt.xlabel("Sale")
plt.show()
```

```python
# More BarPlots

divisions = ["A", "B", "C", "D", "E"]
verkaefe = [70,80,73,65,68]
produkt1 = [60,78,50,40,66]

index = np.arange(5)
width = 0.30

plt.bar(index, verkaefe, width, color='green', label="Seller 1")
plt.bar(index+width, produkt1, width, color='blue', label="Seller 2")

plt.title("Bar Diagramm")
plt.xlabel("Division")
plt.ylabel("sale")
plt.xticks(index+width/2, divisions)
        
plt.legend(loc='best')        
plt.show()
```

```python
# Barplots Examples

divisions = ["Div-A", "Div-B", "Div-C", "Div-D", "Div-E"]
verkaefe = [70,80,73,65,68]
produkt1 = [60,78,50,40,66]
rest = [10, 2, 23, 15, 2]

index = np.arange(5)
width = 0.30

plt.bar(index, produkt1, width, color='green', label="Produkt 1")
plt.bar(index, rest, width, color='red', label="Andere Produkte", bottom=produkt1)

plt.title("Ein gestapeltes Balkendiagramm")
plt.xlabel("Division")
plt.ylabel("Sale")
plt.xticks(index, divisions)
        
plt.legend(loc='best')        
plt.show()
```

```python
# Pie Plots
firmen = ["Firma 1", "Firma 2", "Firma 3", "Firma 4", "Firma 5", "Firma 6"]
handelsvolumen = [80, 50, 112, 68, 90, 17]
Explode = [0,0,0.5,0,0.25,0]
plt.title("Pie Plots!")

#plt.pie(handelsvolumen,labels=firmen)
plt.pie(handelsvolumen, explode=Explode, labels=firmen, startangle=45)
plt.axis("equal")
plt.legend(title="Firmenliste", loc="best")
plt.show()
```

```python
# All together

X_arr = np.arange(10)
Y_arr = 3*X_arr + 2 + np.random.random(size=X_arr.size)
mu, sigma = 15, 1
gaussian_arr = np.random.normal(mu, sigma, size=10000)
bar_arr = np.array(['Frhling', 'Sommer', 'Herbst', 'Winter'])
yerr_arr = np.random.randint(5, 10, 4)
freq_arr = np.random.randint(0, 100, 4)

# Subplots
fig, ax_arr = plt.subplots(nrows=2, ncols=2, sharex=False, sharey=False)

fig.set_figwidth(12)
fig.set_figheight(8)
fig.suptitle("Graphs")

# First
ax_arr[0, 0].scatter(X_arr, Y_arr, marker='+', c='g', s=Y_arr*10) 
ax_arr[0, 0].plot(X_arr, Y_arr, linestyle='dashed', color='k')
ax_arr[0, 0].axis([0, 10, 0, 35])
ax_arr[0, 0].set_title('Scatter upper left')

# Second
ax_arr[0, 1].hist(gaussian_arr, facecolor='g', alpha=0.3)
ax_arr[0, 1].set_title('Histogramm upper right')

# Third
ax_arr[1, 0].boxplot(gaussian_arr, vert=False, showfliers=False, showmeans=True, labels=['Gaussian'])
ax_arr[1, 0].set_title('Box plot')

# Fourth
ax_arr[1,1].bar(bar_arr, freq_arr, yerr = yerr_arr, color='gold')
ax_arr[1, 1].set_title('Bar chart')

plt.show()
```

### Heatmaps

```python
#dataset = load_iris()
dataset = load_wine()

#print(dataset.DESCR)

df = pd.DataFrame(data= np.c_[dataset['data'], dataset['target']], columns= dataset['feature_names'] + ['target'])
df.target = df.target.astype(int)
```

```python
df
```

```python
plt.figure(figsize=(14,5))
sns.heatmap(df.corr(), vmin=-1,cmap='PiYG', annot=True);
plt.yticks(rotation=0)

b, t = plt.ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
plt.ylim(b, t) # update the ylim(bottom, top) values

plt.show()
```
